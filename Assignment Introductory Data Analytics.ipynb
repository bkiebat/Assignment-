```python
def reg_player(df):
    if df in ['3330', '3104', '2995', '2889', '2866', '2831', '2775', '2695', '2570']:
        return 'Regular Player'
    else:
        return "Substitute"
```

```python
goal_stats['Regular Player'] = goal_stats['MATCH TIME'].apply(reg_player)
goal_stats.tail()
```
